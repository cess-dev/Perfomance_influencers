This project investigates the factors influencing academic perfomance among students, it looks into factors such as: attendance, parents education, internet access as well as extra curricular activities.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Student_Performance.csv')
```

```python
# 1. Average Overall Score
avg_overall = df['overall_score'].mean()

# 2. Distribution of Final Grade
grade_counts = df['final_grade'].value_counts()

# 3. Correlation between Math, Science, and English
# This checks if students who do well in one subject do well in others
subject_corr = df[['math_score', 'science_score', 'english_score']].corr()

# 4. School Type Comparison
school_dist = df['school_type'].value_counts()

# 5. Average Study Hours
avg_study = df['study_hours'].mean()

# Output results
print(f"Avg Score: {avg_overall:.2f}")
print(f"Avg Study Hours: {avg_study:.2f}")
print("\nGrade Distribution:\n", grade_counts)
print("\nCorrelation Matrix:\n", subject_corr)
```

Average Performance: The mean overall_score for the 25,000 students is 64.01.
Study Habits: On average, students commit 4.25 study_hours per week.
Core Subject Synergy: There is a strong positive correlation (approx. 0.78) between math_score, science_score, and english_score. This suggests that students who excel in one core area are highly likely to excel in the others.

```python
# Do males and females differ in overall_score?
print(df.groupby('gender')['overall_score'].mean())

# Does one gender perform better in math vs English?
print(df.groupby('gender')[['math_score', 'english_score']].mean())

# Does study_hours differ by gender?
print(df.groupby('gender')['study_hours'].mean())
```

```python
# Do private schools outperform public schools?
print(df.groupby('school_type')['overall_score'].mean())

# Is attendance higher in one school type?
print(df.groupby('school_type')['attendance_percentage'].mean())

# Is internet access more common in one group?
print(pd.crosstab(df['school_type'], df['internet_access'], normalize='index'))
```

```python
# Does parent_education impact overall_score?
print(df.groupby('parent_education')['overall_score'].mean())

# Does it affect study_hours?
print(df.groupby('parent_education')['study_hours'].mean())

# Does it affect final_grade distribution?
print(pd.crosstab(df['parent_education'], df['final_grade'], normalize='index'))
```

Female students generally outperform males in overall_score, math_score, and english_score, though this advantage is closely mirrored by their higher average study_hours. Similarly, students whose parents hold a diploma, post graduate or have no formal education achieve the highest scores, a trend that again correlates directly with their increased time spent studying. Ultimately, the data suggests that study_hours is the primary driver of academic success across all demographic groups, outweighing factors like gender or parental background.

```python
# correlation between hours and scores
correlation = df['study_hours'].corr(df['overall_score'])
print(f"Correlation between Study Hours and Overall Score: {correlation:.4f}")

# Check for Diminishing Returns
# Rounding hours to the nearest whole number to see the trend more clearly
df['hours_rounded'] = df['study_hours'].round()
diminishing_returns = df.groupby('hours_rounded')['overall_score'].mean()
print("\nScore by Study Hours (Trend):")
print(diminishing_returns)

# Effectiveness by Study Method
study_method_perf = df.groupby('study_method')['overall_score'].mean()
print("\nPerformance by Study Method:")
print(study_method_perf)

# Hours spent on each study method
study_method_hours = df.groupby('study_method')['study_hours'].mean()
print(study_method_hours)
```

```python
# Attendance vs. Study Hours Predictability
# Comparing which has a higher correlation with overall_score
attendance_corr = df['attendance_percentage'].corr(df['overall_score'])
study_corr = df['study_hours'].corr(df['overall_score'])
print(f"Correlation (Attendance vs Score): {attendance_corr:.4f}")
print(f"Correlation (Study Hours vs Score): {study_corr:.4f}")

# What happens below 75% attendance?
low_attendance_avg = df[df['attendance_percentage'] < 75]['overall_score'].mean()
high_attendance_avg = df[df['attendance_percentage'] >= 75]['overall_score'].mean()
print(f"\nAvg Score (Attendance < 75%): {low_attendance_avg:.2f}")
print(f"Avg Score (Attendance >= 75%): {high_attendance_avg:.2f}")
```

```python
# Checking if internet access correlates with higher scores or acts as a distraction
internet_impact = df.groupby('internet_access')['overall_score'].mean()
print("\nImpact of Internet Access on Scores:")
print(internet_impact)
```

```python
# Checking if long commutes "steal" time from studying or reduce attendance
travel_impact = df.groupby('travel_time')[['overall_score', 'study_hours', 'attendance_percentage']].mean()
print("\nImpact of Travel Time:")
print(travel_impact)
```
